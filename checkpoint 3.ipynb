```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
print(plot_correlation_map(df))    
```

### This function gives us a clear analysis of each column and its relation to all the others so it is clearer to be able to find the link between each column. This way we can clearly see that 'Age' and 'Sex' have a direct impact on the column 'Survivor' while others like 'Fare' and 'SibSp' have no significant impact on the number of survivors. Hence, we can now drop the columns that don't have any impact or function in such a way as simplifying the Data Frame
